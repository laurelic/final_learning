```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read with pandas
sqlitedbpath = "../db/inpatient.db"
conn = sqlite3.connect(sqlitedbpath)
# linelimit = 129400; # set this for testing, remove later
# df = pd.read_sql_query(f"select * from inpatient limit {linelimit};", conn)
df = pd.read_sql_query("select * from inpatient;", conn)
conn.close()
print(len(df))
print(df.columns)
print(df.dtypes)
```

```python
# remove $ and , from the currency columns
# row 129340 has no '$' for the payment columns => cast x to str to avoid errors
df.average_covered_charges = df.average_covered_charges.apply(lambda x: str(x).replace('$','').replace(',',''))
df.average_total_payments = df.average_total_payments.apply(lambda x: str(x).replace('$','').replace(',',''))
df.average_medicare_payments = df.average_medicare_payments.apply(lambda x: str(x).replace('$','').replace(',',''))
```

```python
numcols=['total_discharges', 'average_covered_charges', 'average_total_payments','average_medicare_payments']
df[numcols] = df[numcols].apply(pd.to_numeric, errors='coerce')
print(len(df), df.columns, df.dtypes)
```

```python
# cast procedure id to numeric for now
df['procedure_id'] = pd.to_numeric(df["drg_definition"].str.split(" -", n = 1, expand = True)[0])
```

```python
select_fraction=0.1 # select only a fraction of data for now
select_columns=['procedure_id', 'provider_id', 'provider_zip_code', 'total_discharges', 'average_covered_charges', 'average_total_payments', 'average_medicare_payments']
inpatient = df[select_columns].sample(frac=select_fraction, replace=False)
print(len(inpatient))
```

```python
print(inpatient.shape)
print(inpatient.columns)
print(inpatient.dtypes)
inpatient.describe()
inpatient['procedure_id'].hist(bins=50)
```

```python
inpatient.plot.scatter(x = 'total_discharges',
                       y = 'average_covered_charges',
                       c = 'procedure_id',
                       colormap='viridis')
```

```python
#plt.style.use('ggplot')
scmatrix = pd.plotting.scatter_matrix(inpatient, figsize=(12,12),
                             diagonal='hist',
                             marker='o',
                             grid=True)
plt.show()
```
